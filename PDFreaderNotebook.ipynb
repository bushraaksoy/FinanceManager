## ***Kaspi Bank* Statement Reader**

```python
import pdfplumber

statement = []

with pdfplumber.open("/content/drive/MyDrive/FM/kaspi_gold_statement.pdf") as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        statement.append(text)

```

```python
statement
```

```python
statement = statement[0].split("\n")
```

```python
for line in statement:
  if "number" in line:
    statement.remove(line)
```

```python
statement
```

```python
import re
period = []
for line in statement:
    if ("period" in line) or ("Period" in line):
      dates = re.findall(r"\d{2}\.\d{2}\.\d{2}", line)
      period.append(dates)
```

```python
start_date = period[0][0]
end_date = period[0][1]
```

```python
index = next((i for i, s in enumerate(statement) if "Date" in s or "date" in s), -1)
```

```python
transaction_history = statement[index:]
transaction_history
```

```python
transactions = []
transaction_pattern = r"(\d{2}\.\d{2}\.\d{2})\s+([\+\-]\s[\d\s,]+ ₸)\s+([\w\s\"().,]+)"
for line in transaction_history:
  if len(re.findall(transaction_pattern, line))!=0:
      matches = re.findall(transaction_pattern, line)
      transactions.append(matches)

```

```python
print(transactions)
```

```python
transaction_history = pd.DataFrame([item[0] for item in transactions], columns=["Date", "Amount", "Transaction Details"])

transaction_history
```

```python
transaction_history[['Transaction Type', 'Description']] = transaction_history['Transaction Details'].apply(
    lambda x: pd.Series(x.split(' ', 1)) if ' ' in x else pd.Series([x, ''])
)

# Удаляем старый столбец 'Transaction Details'
transaction_history.drop(columns=['Transaction Details'], inplace=True)
```

## ***Halyk Bank* Statement Reader**

```python
import pdfplumber

halyk_statement = []

with pdfplumber.open("/content/drive/MyDrive/FM/pdf_statements_halyk.pdf") as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        halyk_statement.append(text)
    print(len(pdf.pages))
```

```python

import pdfplumber

transactions_tables = []  # list of tables

with pdfplumber.open("/content/drive/MyDrive/FM/pdf_statements_halyk.pdf") as pdf:
    for page in pdf.pages:
        tables = page.extract_tables()
        for table in tables:
            transactions_tables.append(table)  #adding each table as list
```

```python
import re

first_date_re = r"\d{2}\.\d{2}\.\d{4}"
first_date_indices = []

for table in transactions_tables:
    index_of_first_date = next(
        (i for i, row in enumerate(table) if any(re.search(first_date_re, str(cell)) for cell in row)),
        None
    )
    first_date_indices.append(index_of_first_date)

print(first_date_indices)
```

```python
columns = transactions_tables[0][0]
columns
```

```python
columns = [name.replace("\n", " ") for name in columns]
columns
```

```python
clean_tables = []
for table, first_date_index in zip(transactions_tables, first_date_indices):
    clean_table = table[first_date_index:]
    clean_tables.append(clean_table)
```

```python
# len(transactions_tables)
len(clean_tables)
```

```python
halyk_transactions = pd.concat([pd.DataFrame(table, columns=columns) for table in clean_tables], ignore_index=True)
halyk_transactions = halyk_transactions.drop(halyk_transactions.columns[-1], axis=1) #deleting card number column
halyk_transactions
```

```python
halyk_statement = halyk_statement[0].split("\n")
```

```python
import re
halyk_peroid = []
for line in halyk_statement:
    if ("period" in line) or ("Period" in line):
      dates = re.findall(r"\d{2}\.\d{2}\.\d{4}", line)
      halyk_peroid.append(dates)
```
